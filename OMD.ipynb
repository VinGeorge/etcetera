```python
import pandas as pd
import chardet
```

##### Для демонстрации решения проверял на файле [6; 17], но функция будет работать для любого количества колонок и строк

```python
def get_boolean(file_csv, min_value, max_value):
    headers = ['id']
    for i in range(min_value, max_value):
        header = 'Var_{}'.format(i)
        headers.append(header)
        
    name = file_csv
    with open(file_csv, 'rb') as f:
        read = f.read()
        detect = chardet.detect(read)
        decoding = detect['encoding']
        
    file = pd.read_csv(file_csv, sep ='\s+', encoding = decoding, na_values = True, names=headers)
    
    test_pd = file.notnull()
    
    test_finish = test_pd.replace(to_replace = [True, False], value = [1, 0])
    
    ##Предполагаю, что можно использовать колонку идексов pd вместо номера строки в исходном файле, поэтому ее удаляю, но нужно смотреть на то, как массив будет использоваться дальше
    del test_finish['id'] 
    
    return test_finish
    
```
